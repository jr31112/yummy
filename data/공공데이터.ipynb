## 관광지 모델에 끼워맞추기
```python
# 4 Spot(관광지)
class Spot(models.Model):
    id = models.IntegerField(primary_key=True) #auto increment
    spot_name = models.CharField(max_length=50)
    road_address  = models.CharField(max_length=200, null=True,  blank=True)
    address= models.CharField(max_length=200, null=True,  blank=True)
    latitude = models.FloatField(max_length=10, null=True,  blank=True)
    longitude = models.FloatField(max_length=10, null=True,  blank=True)
    description = models.TextField()
```

````python
df1.columns
````

````python
df1[['공공편익시설정보','숙박시설정보','운동및오락시설정보','휴양및문화시설정보','접객시설정보','지원시설정보','관리기관명','관리기관전화번호']] = df1[['공공편익시설정보','숙박시설정보','운동및오락시설정보','휴양및문화시설정보','접객시설정보','지원시설정보','관리기관명','관리기관전화번호']].fillna("없음")
````

````python
def makeDescription(row):

    description = ""
    if row['관광지소개'] !="없음":
        description+="소개 : "+row['관광지소개']+"\n"
    if row['공공편익시설정보'] !="없음":
        description+="공공편익시설정보 : "+row['공공편익시설정보']+"\n"
    if row['숙박시설정보']  !="없음":
        description+="숙박시설정보 : "+row['숙박시설정보']+"\n"
    if row['운동및오락시설정보'] !="없음":
        description+="운동및오락시설정보 : "+row['운동및오락시설정보']+"\n"
    if row['휴양및문화시설정보']  !="없음":
        description+="휴양및문화시설정보 : "+row['휴양및문화시설정보']+"\n"
    if row['접객시설정보']  !="없음":
        description+="접객시설정보 : "+row['접객시설정보']+"\n"
    if row['지원시설정보']  !="없음":
        description+="지원시설정보 : "+row['지원시설정보']+"\n"
    if row['관리기관명']  !="없음":
        description+="관리기관명 : "+row['관리기관명']+"\n"
    if row['관리기관전화번호']  !="없음":
        description+="관리기관전화번호 : "+row['관리기관전화번호']+"\n"
        
    return description
        
df1['description'] =df1.apply(makeDescription,axis=1)
````

````python
df1
````

````python
df1_result = df1[['관광지명','소재지도로명주소','소재지지번주소','위도','경도','description']]
df1_result
````

````python
df1_result = df1_result.rename(columns={'관광지명':'spot_name','소재지도로명주소':'road_address','소재지지번주소':'address','위도':'latitude','경도':'longitude'})
````

````python
df2[['부대시설','주차장보유여부','결제방법','홈페이지주소','주변관광정보','전화번호']] = df2[['부대시설','주차장보유여부','결제방법','홈페이지주소','주변관광정보','전화번호']].fillna("없음")
````

````python
def makeDescription1(row):

    description = ""
    if row['부대시설'] !="없음":
        description+="부대시설 : "+row['부대시설']+"\n"
    if row['주차장보유여부']  !="없음":
        description+="주차장보유여부 : "+row['주차장보유여부']+"\n"
    if row['결제방법'] !="없음":
        description+="결제방법 : "+row['결제방법']+"\n"
    if row['홈페이지주소']  !="없음":
        description+="홈페이지주소 : "+row['홈페이지주소']+"\n"
    if row['주변관광정보']  !="없음":
        description+="주변관광정보 : "+row['주변관광정보']+"\n"
    if row['전화번호']  !="없음":
        description+="전화번호 : "+row['전화번호']+"\n"
   
        
    return description
        
df2['description'] =df2.apply(makeDescription1,axis=1)
````
